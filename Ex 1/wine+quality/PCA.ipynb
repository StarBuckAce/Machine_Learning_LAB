```python
data_mean = np.mean(X, axis=0)
standardized_data = X- data_mean
print("\n Standarized Dataset:- ")
print(standardized_data)
```

```python
cov_matrix = np.cov(standardized_data, rowvar=False)
print("\nCovariance Matrix")
print(cov_matrix)
```

```python
eigenvalues,eigenvectors = np.linalg.eig(cov_matrix)
print("\n Eigenvalues")
print(eigenvalues)
print("\n Eigenvectors")
print(eigenvectors)
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
top_eigenvectors = eigenvectors[:, :2]
```

```python
pca_data = np.dot(standardized_data, top_eigenvectors)
print("\nPCA Reduced Data:")
print(pca_data)
```

```python
# Step 8: Explained Variance Ratio
explained_variance_ratio = eigenvalues[:2] / sum(eigenvalues)
print("\nExplained Variance Ratio for top 2 components:")
print(explained_variance_ratio)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], color='blue', edgecolor='k', s=100)

plt.title("PCA Result (2 Components)", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=12)
plt.ylabel("Principal Component 2", fontsize=12)
plt.grid(True)
plt.show()
```

```python
top_eigenvectors = eigenvectors[:, :3]
```

```python
pca_data = np.dot(standardized_data, top_eigenvectors)
print("\nPCA Reduced Data:")
print(pca_data)
```

```python
# Step 8: Explained Variance Ratio
explained_variance_ratio = eigenvalues[:3] / sum(eigenvalues)
print("\nExplained Variance Ratio for top 2 components:")
print(explained_variance_ratio)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Assuming 'data' is your pandas DataFrame containing the numerical features
# 1. Standardize the data
x = StandardScaler().fit_transform(df) 

# 2. Apply PCA
pca = PCA(n_components=2)  # Choose the desired number of components
principalComponents = pca.fit_transform(x)

# 3. Create a DataFrame for the principal components
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'], c=principalDf['principal component 1'], cmap='viridis') 
# Color based on PC1 values using 'viridis' colormap

plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('Scatter Plot of PC1 vs. PC2')
plt.colorbar(label='PC1 Value')  # Add a colorbar to show the color scale
plt.grid(True)  # Add a grid for better visualization
plt.show()
```

```python
pca = PCA(n_components=3)  # Choose the desired number of components
principalComponents = pca.fit_transform(x)

# 3. Create a DataFrame for the principal components
principalDf1 = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2','principal component 3'])




from mpl_toolkits.mplot3d import Axes3D

# Assuming 'principalDf' contains the principal components
fig = plt.figure(figsize=(10, 8))  # Adjust figure size if needed
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

# Create 3D scatter plot
ax.scatter(principalDf1['principal component 1'], principalDf1['principal component 2'], principalDf1['principal component 3'], 
           c=principalDf1['principal component 1'], cmap='viridis')  # Color based on PC1 values

ax.set_xlabel('Principal Component 1 (PC1)')
ax.set_ylabel('Principal Component 2 (PC2)')
ax.set_zlabel('Principal Component 3 (PC3)')
ax.set_title('3D Scatter Plot of PC1, PC2, and PC3')
plt.show()
```

```python
pca = PCA()  # Initialize PCA without specifying the number of components
pca.fit(x)  # Fit PCA to the data

# Get the explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_

# Plot the elbow plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.plot(range(1, len(explained_variance_ratios) + 1), explained_variance_ratios, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Elbow Plot for PCA')
plt.grid(True)  # Add a grid for better visualization
plt.show()
```

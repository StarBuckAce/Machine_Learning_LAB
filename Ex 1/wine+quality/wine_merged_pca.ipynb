```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
white = pd.read_csv('winequality-white.csv',delimiter = ';')
white
```

```python
red = pd.read_csv('winequality-red.csv', delimiter = ';')
red
```

```python
combined = [white,red]
df = pd.concat(combined, ignore_index = True)
df['classification'] = [1] * len(white) + [0] * len(red)
```

```python
x1 = df['fixed acidity'].mean()
x2 = df['volatile acidity'].mean()
x3 = df['citric acid'].mean()
x4 = df['residual sugar'].mean()
x5 = df['chlorides'].mean()
x6 = df['free sulfur dioxide'].mean()
x7 = df['total sulfur dioxide'].mean()
x8 = df['density'].mean()
x9 = df['pH'].mean()
x10 = df['sulphates'].mean()
x11 = df['alcohol'].mean()
```

```python
print("Mean of fixed acidity:", x1)
print("Mean of volatile acidity:", x2)
print("Mean of citric acid:", x3)
print("Mean of residual sugar:", x4)
print("Mean of chlorides:", x5)
print("Mean of free sulfur dioxide:", x6)
print("Mean of total sulfur dioxide:", x7)
print("Mean of density:", x8)
print("Mean of pH:", x9)
print("Mean of sulphates:", x10)
print("Mean of alcohol:", x11)

x_mean = [x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11]
            
cov_matrix = np.zeros((11, 11))

#covariance matrix
for i in range(11):
    for j in range(11):
        if i == j:
            cov_matrix[i, j] = np.sum((df.iloc[:, i] - x_mean[i]) ** 2) / (len(df) - 1)
        else:
            cov_matrix[i, j] = np.sum((df.iloc[:, i] - x_mean[i]) * (df.iloc[:, j] - x_mean[j])) / (len(df) - 1)

print("\n\nCovariance Matrix:\n")
print(cov_matrix)

#eigenvalues and eigenvectors of the covariance matrix
eigenvalues,eigenvectors = np.linalg.eig(cov_matrix)

idx = np.argsort(-eigenvalues)
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
 
print("\n\nEigenvalues:\n\n", eigenvalues)
print("\n\nEigenvectors:\n")
print(eigenvectors)

k = 3
pc1 = eigenvectors[:, 0]
pc2 = eigenvectors[:, 1]
pc3 = eigenvectors[:, 2]

print("\n\nFirst principal component :\n\n", pc1)
print("\n\nSecond principal component :\n\n", pc2)
print("\n\nThird principal component :\n\n", pc3)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

pc1_scores = df.iloc[:, :11].values @ pc1
pc2_scores = df.iloc[:, :11].values @ pc2
pc3_scores = df.iloc[:, :11].values @ pc3

set1 = df[df['classification'] == 0]
set2 = df[df['classification'] == 1]

ax.scatter(pc1_scores[set1.index], pc2_scores[set1.index], pc3_scores[set1.index], color='r', label='Class 1')  
ax.scatter(pc1_scores[set2.index], pc2_scores[set2.index], pc3_scores[set2.index], color='g', label='Class 2')

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.title('Data Points in the First Three Principal Components')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pc1_scores[set1.index], pc2_scores[set1.index], color='r', label='Class 1', alpha=0.6)
plt.scatter(pc1_scores[set2.index], pc2_scores[set2.index], color='g', label='Class 2', alpha=0.6)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Data Points in the First Two Principal Components')
plt.legend()
plt.grid()
plt.show()
```
